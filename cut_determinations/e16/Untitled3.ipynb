```python
def nrand():
    return (random.randrange(0,2000,1)/1000.0)-1.0

def nrand2():
    return (random.randrange(0,2000,1)/2000.0)

def f(x_, a_, b_, c_):
    return a_* x_*x_ + b_*x_ + c_

def Gaus(x_, a_, m_, s_):
    return a_*np.exp(((x_-m_)**2.0)/(2.0*s_*s_))
```

```python
x= [i for i in range(29)]
a=-0.5
b=3.3
c=6.2

s=[0.1,0.2]

e=[0.2,0.3]
y0 = [f(x[i],a,b,c) for i in range(29)]
y1 = [f(x[i],a,b,c)*(1.0+s[0]*nrand()) for i in range(29)]
y2 = [f(x[i],a,b,c)*(1.0+s[1]*nrand()) for i in range(29)]

y0err = [np.sqrt(np.abs(y0[i])) for i in range(29)]
y1err = [np.sqrt(np.abs(y0[i]+e[0]*nrand2())) for i in range(29)]
y2err = [np.sqrt(np.abs(y0[i]+e[1]*nrand2())) for i in range(29)]

sys_err = [np.std([y0[i],y1[i],y2[i]]) for i in range(29)]
rel_sys_err = [sys_err[i]/np.abs(y0[i]) for i in range(29)]



plt.scatter(x,y0)
plt.errorbar(x,y0,yerr=y0err,ls='')
plt.scatter(x,y1)
plt.errorbar(x,y1,yerr=y1err,ls='')
plt.scatter(x,y2)
plt.errorbar(x,y2,yerr=y2err,ls='')
plt.bar(x,sys_err)
plt.show()

plt.bar(x,rel_sys_err)
plt.show()
```
